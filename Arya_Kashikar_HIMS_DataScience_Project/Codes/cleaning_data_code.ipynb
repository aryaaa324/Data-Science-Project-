```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Load Dataset
df = pd.read_csv("C://Users//91971//Desktop//HIMS_DataScience_Project//dataset//aryakashikar03_17536107524728932.csv")
```

```python
# Initial Overview
print("Shape:", df.shape)
print("Preview:\n", df.head())
```

```python
# 2. Data Cleaning

# Clean column names
df.columns = df.columns.str.strip().str.replace(" ", "_")
df.columns
```

```python
#check sum of missing values 
df.isnull().sum()
```

```python
df = df.dropna(subset=['State', 'District', 'Sub-District', 'Year'])  # drop only if these are missing
```

```python
# Drop columns where more than 90% of the values are missing
threshold = 0.9
df = df.loc[:, df.isnull().mean() < threshold]
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
# 1. Check for duplicate rows
duplicate_rows = df[df.duplicated()]
print(f"ðŸ” Duplicate rows found: {duplicate_rows.shape[0]}")

```

```python

# 3. Check for duplicate columns
duplicate_columns = df.columns[df.columns.duplicated()].tolist()
print(f"ðŸ“› Duplicate columns found: {duplicate_columns}")
```

```python
print("Final shape after full cleaning:", df.shape)
print("Remaining missing values:", df.isnull().sum().sum())
```

```python
# 6. Save cleaned data to CSV
output_path = "C://Users//91971//Desktop//HIMS_DataScience_Project//dataset//HMIS_Cleaned.csv"
df.to_csv(output_path, index=False)

print("âœ… File saved successfully at:", output_path)
```
